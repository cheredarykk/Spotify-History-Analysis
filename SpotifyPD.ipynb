Import Pandas Library

```python
import pandas as pd
```

Load the CSV File using (read_csv) method and save it as dataframe (sphdf)

```python
sphdf = pd.read_csv(r"C:\\Users\Anna\Documents\DA\Projects\Spotify\my_streaming_history_csv.csv", on_bad_lines='skip', delimiter=';')
sphdf.head()
```

Check for missing values

```python
sphdf.isnull().sum()
```

```python
sphdf = sphdf.where(pd.notnull(sphdf), 0)
```

```python
sphdf.isnull().sum()
```

Check Data Types

```python
sphdf.dtypes
```

```python
sphdf['Date']= pd.to_datetime(sphdf['Date'], format="%d.%m.%Y")

sphdf['EndTime']= pd.to_datetime(sphdf['EndTime'], format="%H:%M:%S")
```

Delete unnecesary columns

```python
del sphdf["minutes"]
```

```python
del sphdf["hours "]
```

DataFrame Export

Convert all rows into a list of tuples

```python
y = []

for i in range (len(sphdf)):
    x = tuple(sphdf.iloc[i])
    y.append(x)
    
y
```

Save the list as a txt file

```python
file = open('streaming_history.txt', 'w')

for tuple in y:
    file.write(str(tuple)+','+ '\n')
    
file.close()
```
